```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('h_1_g_1000', 'rb') as fp:
    h_1 = pickle.load(fp)
    
with open('h_2_g_1000', 'rb') as fp:
    h_2 = pickle.load(fp)
```

```python

plt.plot(np.arange(1000), h_1, np.arange(1000), h_2)
plt.xlabel('Number of Graphs')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve')
plt.show()
```
